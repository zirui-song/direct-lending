```python
import wrds
import pandas as pd

# Step 1: Connect to WRDS
db = wrds.Connection(wrds_username='zrsong')
# Step 2: Create a list of accession numbers
accession_numbers = [
    '0001393066-16-000026',  # Example accession numbers
    '0001193125-19-278422',
    '0001504337-19-000028'
    # Add your own list of accession numbers here
]

# Convert accession numbers list into a format suitable for SQL query
accession_str = "','".join(accession_numbers)

# Step 3: Query the SEC Analytics Suite database to retrieve filings based on accession numbers
query = f"""
    SELECT
        filing_view.accession,
        filing_view.form,
        filing_view.filing_date,
        registrant.cik,
        filing_view.filing
    FROM
        wrds_sec_search.filing_view
    JOIN
        wrds_sec_search.registrant ON registrant.accession = filing_view.accession
    WHERE
        filing_view.accession IN ('{accession_str}')
"""

# Step 4: Execute the query and store results in a DataFrame
filings_df = db.raw_sql(query)

# Step 5: Close the WRDS connection when done
db.close()

# Step 6: Save the result as a CSV file or process the filing texts further
filings_df.to_csv("../Data/filings_data.csv", index=False)

# Alternatively, display the filings data
print(filings_df)
```
